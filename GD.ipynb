```
import numpy as np
import matplotlib.pyplot as plt
```

```
#we create a numpy 1D array 
x=np.array([1,2,3,4,5])  #independent means input variable
y=np.array([5,7,9,11,13]) #target variable means dependent variable
```

```
print(x)
```

```
print(y)
```

```
#create function for gradient descent
def gradient_descent(x,y):
    b_old=0  #suppose assign initial value of intercept =0 in ML but DL b=bias
    m_old=0 # suppose assign initial value of coefficient/slope=0 but DL m=weight
    n=len(x) # No. of datapoints means records means rows find  in dataset
    iterations=910 #suppose decide no. of iterations=1000 in ML but epoches in DL
    learning_rate=0.01 #or assign 0.05 or 0.1 it is also called lambda
    print("slope m =",m_old," Intercept b =",b_old)
    plt.scatter(x,y,color="red")
    for i in range(iterations): #start=0 stop=1000-1=999 step=+1
      Y_pred=m_old*x+b_old  #Y=mX+b
      plt.plot(x,Y_pred,color='red')#draw line graph
      cost=(1/n)*sum([val**2 for val in (y-Y_pred)]) #cost means mse 
      #minimize the cost or mse means error then we do diffrentiate 
      #wrt slope(thetha1) and wrt intercept(thetha0)
      mdiff=-(2/n)*sum(x*(y-Y_pred))
      bdiff=-(2/n)*sum(y-Y_pred)
      m_old=m_old-learning_rate*mdiff
      b_old=b_old-learning_rate*bdiff
      
      print("Slope : m :",m_old,"Intercept b : ",b_old,"Iteration : ", i," MSE : ",cost)
    return (Y_pred) #return final predicted value if achieve target
    #means min mse

    
```

```
#call in main program
Y_pred=gradient_descent(x,y)
print("Predicted Output : ",Y_pred)
```
